```python
#1. Scrape the details of most viewed videos on YouTube from Wikipedia. Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos
# You need to find following details:A) Rank B) Name C) Artist D) Upload date E) Views

#connect to web driver
chrome_options=Options()
chrome_options.add_argument("disable--infobars")
chrome_options.add_argument("--incognito")
driver=webdriver.Chrome(r"C:\Users\shaha\Desktop\My Course\DataTrained\chromedriver_win32\chromedriver.exe", options=chrome_options)

# webpage by url :https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos
url="https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos" 
driver.get(url)

#maximize the chrome window
driver.maximize_window() 
```

```python
# 6. Scrape the details of top 100 songs on billiboard.com.Url = https:/www.billboard.com/  You have to find the following details:
#A) Song name B) Artist name C) Last week rank D) Peak rank E) Weeks on board Note: - From the home page you have to click on the charts option then hot 100-page link through code.

#connect to web driver
chrome_options=Options()
chrome_options.add_argument("disable--infobars")
chrome_options.add_argument("--incognito")
driver=webdriver.Chrome(r"C:\Users\shaha\Desktop\My Course\DataTrained\chromedriver_win32\chromedriver.exe", options=chrome_options)

# webpage by url :
url="https:/www.billboard.com/" 
driver.get(url)
delay = 30  #wait till we find this xpath
WebDriverWait(driver,delay).until(EC.element_to_be_clickable((By.XPATH,"//div[@class='main-sub-menu // u-padding-tb-075@desktop u-display-none@mobile-max lrv-u-width-100p u-box-shadow-header-main lrv-u-position-relative u-z-index-middle-bottom']/nav[1]/ul/li[1]/a")))
charts = driver.find_elements_by_xpath("//div[@class='main-sub-menu // u-padding-tb-075@desktop u-display-none@mobile-max lrv-u-width-100p u-box-shadow-header-main lrv-u-position-relative u-z-index-middle-bottom']/nav[1]/ul/li[1]/a")
for i in charts:
    link = i.get_attribute('href')
    if 'charts/hot-100' in link:  #getting link of reuired page
        driver.implicitly_wait(10)
        driver.get(link)
        break
        
        
#maximize the chrome window
driver.maximize_window()
```
