```python
#1. Write a python program which searches all the product under a particular product from www.amazon.in. 
#The product to be searched will be taken as input from user. For e.g. If user input is ‘guitar’. Then search for guitars.

#connect to web driver
chrome_options=Options()
chrome_options.add_argument("--incognito")
service=Service(r"C:\Users\shaha\Desktop\My Course\DataTrained\chromedriver_win32\chromedriver.exe")
driver = webdriver.Chrome(service=service, options=chrome_options)
driver.get('https://www.amazon.in/')
driver.maximize_window()
time.sleep(2)
```

```python
#Entering “products” in search field in input method 
search_field=driver.find_element_by_xpath("//div//div[@class='nav-search-field ']/input[@id='twotabsearchtextbox']")
search_field.send_keys(input('product ',))

#selecting perticular product from dropdown
s=driver.find_element_by_id("searchDropdownBox")
drop=Select(s)
drop.select_by_value('search-alias=mi')
time.sleep(2)


# clicking on search button.
search_button=driver.find_element_by_xpath("//div[@class='nav-search-submit nav-sprite']") 
search_button.click()
```

```python
#2.In the above question, now scrape the following details of each product listed in first 3 pages of your search results and save it in a data frame and csv. In case if any product has less than 3 pages in search 
# results then scrape all the products available under that product name. Details to be scraped are: "Brand Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability" and 
# “Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“.

brand_name=[]
product=[]
price=[]
exp_delivary=[]
avl=[]
url=[]
re=[]

driver.refresh()
```

```python
for i in range(0,3):
    try:
        for a in driver.find_elements_by_xpath("//div/div/h2/a[@class='a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal']"):
            url.append(a.get_attribute('href'))
    except Exception as e:
        print("Failed", e)
        
        next_button=driver.find_element_by_xpath('//a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]').click()
        time.sleep(5)   
    
```

```python
for i in url:
    try:
        driver.get(i)
        driver.implicitly_wait(5)
    except:
        driver.implicitly_wait(5)
        continue
    for a in driver.find_elements_by_id("productTitle"):
        brand_name.append(a.text.split(' ')[0])   
    else:
        brand_name.append('-')
        driver.implicitly_wait(1)
    for a in driver.find_elements_by_id("productTitle"):
            product.append(a.text)   
    else:
        product.append('-')
        driver.implicitly_wait(1)
    for a in driver.find_elements_by_id('corePrice_feature_div'):
        price.append(a.text.replace('\n','.'))
    else:
        price.append('-')
        driver.implicitly_wait(1)
    try:
        return_ex=driver.find_element_by_xpath("//div[@class='a-row icon-farm-wrapper']").find_elements_by_xpath(".//div")
        found=False
        for i in return_ex:
            if(i.get_attribute('data-name')=='RETURNS_POLICY'):
                found=True
                re.append(i.find_element_by_xpath(".//span[1]/div[2]/a").text)
            if(found==False):
                re.append('-')         
    except:
        re.append('-')
        driver.implicitly_wait(1)
    try:
        for a in driver.find_elements_by_id('ftCountdown'):
            exp_delivary.append(a.text)
    except:
        exp_delivary.append('-')
        driver.implicitly_wait(1)
    try:
        for a in driver.find_elements_by_xpath('//span[@class="a-size-medium a-color-success"]'):
            avl.append(a.text)
    except:
        avl.append('-')
        driver.implicitly_wait(1)
        continue
```

```python
details=pd.DataFrame({'Brand Name':brand_name[:59],'Name of the Product':product[:59],'Price':price[:59],'Return/Exchange':re[:59],'Expected Delivery':exp_delivary[:59],'Availability':avl[:59],'Product URL':url[:59]})

details
```

```python
# Q4 Write a python program to search for a smartphone(e.g.: Oneplus Nord, pixel 4A, etc.) on www.flipkart.com and scrape following details for all the search results displayed on 1st page. Details to be 
#scraped: “Brand Name”, “Smartphone name”, “Colour”, “RAM”, “Storage(ROM)”, “Primary Camera”,“Secondary Camera”, “Display Size”, “Battery Capacity”, “Price”, “Product URL”. Incase if any of the 
# details is missing then replace it by “- “. Save your results in a dataframe and CSV.

#connect to web driver
chrome_options=Options()
chrome_options.add_argument("disable--infobars")
chrome_options.add_argument("--incognito")
driver=webdriver.Chrome(r"C:\Users\shaha\Desktop\My Course\DataTrained\chromedriver_win32\chromedriver.exe", options=chrome_options)

# Flipkart webpage by url : https://www.flipkart.com/
url="https://www.flipkart.com/" 
driver.get(url)

#maximize the chrome window
driver.maximize_window() 

# closing login pop-up window
Close_button=driver.find_element_by_xpath('/html/body/div[2]/div/div/button') 
Close_button.click()

#Entering “Oneplus Nord, pixel 4A” in search field where “search for products, brands and more” 
search_field=driver.find_element_by_class_name("_3704LK ") 
search_field.send_keys("Oneplus Nord, pixel 4A")

# clicking on search button.
search_button=driver.find_element_by_class_name('_34RNph') 
search_button.click()
time.sleep(2)

#Creating empty lits
Brand=[]
name=[]
color=[]
ram=[]
rom=[]
cam1=[]
cam2=[]
Display_Size=[]
Battery_Capacity=[]
url=[]
Price=[]

try:
    for a in driver.find_elements_by_xpath("//a[@class='_1fQZEK']"):
        url.append(a.get_attribute('href'))
except Exception as e:
    print("Failed", e)
```

```python
for i in url:
    try:
        driver.get(i)
        driver.implicitly_wait(2)
    except:
        driver.implicitly_wait(2)
        continue
    for a in driver.find_elements_by_xpath("//span[@class='B_NuCI']"):
        Brand.append(a.text[:11])   
    else:
        Brand.append('-')
        driver.implicitly_wait(1)
    for a in driver.find_elements_by_xpath("//span[@class='B_NuCI']"):
        name.append(a.text)   
    else:
        name.append('-')
        driver.implicitly_wait(1)
    for a in driver.find_elements_by_xpath("//div[@class='_4rR01T']"):
        color.append(a.text)   
    else:
        color.append('-')
        driver.implicitly_wait(1)        
    for a in driver.find_elements_by_xpath("//div[@class='_2418kt']/ul/li[1]"):
        ram.append(a.text[:10])
    else:
        ram.append('-')
        driver.implicitly_wait(1)
    for a in driver.find_elements_by_xpath("//div[@class='_2418kt']/ul/li[1]"):
        rom.append(a.text[11:])
    else:
        rom.append('-')
        driver.implicitly_wait(1)
    for a in driver.find_elements_by_xpath("//div[@class='_2418kt']/ul/li[3]"):
        cam1.append(a.text)
    else:
        cam1.append('-')
        driver.implicitly_wait(1)

    for a in driver.find_elements_by_xpath("//div[@class='_1mXcCf']"):
        cam2.append(a.text[100:150])
    else:
        cam2.append('-')
        driver.implicitly_wait(1)               
        
    for a in driver.find_elements_by_xpath("//div[@class='_2418kt']/ul/li[2]"):
        Display_Size.append(a.text)
    else:
        Display_Size.append('-')
        driver.implicitly_wait(1)
    for a in driver.find_elements_by_xpath("//div[@class='_2418kt']/ul/li[4]"):
        Battery_Capacity.append(a.text)
    else:
        Battery_Capacity.append('-')
        driver.implicitly_wait(1)
    for a in driver.find_elements_by_xpath("//div[@class='_30jeq3 _16Jk6d']"):
        Price.append(a.text)
    else:
        Price.append('-')
        driver.implicitly_wait(1)
```

```python
#create DataFrame & save to CSV file
details=pd.DataFrame({'Brand Name':Brand[:24],'Name of the Product':name[:24],'Price':Price[:24],'Color':color[:24],'RAM':ram[:24],'ROM':rom[:24],'Product URL':url[:24],'Primary Camera':cam1[:24],'Secondary Camera':cam2[:24],'Display_Size':Display_Size[:24],
                     'Battery_Capacity':Battery_Capacity[:24]})

details.to_csv(r'C:\Users\shaha\Desktop\My Course\SMPhone.csv')

details
```
