```python
for i in url:
    try:
        driver.get(i)
        driver.implicitly_wait(5)
    except:
        driver.implicitly_wait(5)
        continue
    for a in driver.find_elements_by_id("productTitle"):
        brand_name.append(a.text.split(' ')[0])   
    else:
        brand_name.append('-')
        driver.implicitly_wait(1)
    for a in driver.find_elements_by_id("productTitle"):
            product.append(a.text)   
    else:
        product.append('-')
        driver.implicitly_wait(1)
    for a in driver.find_elements_by_id('corePrice_feature_div'):
        price.append(a.text.replace('\n','.'))
    else:
        price.append('-')
        driver.implicitly_wait(1)
    try:
        return_ex=driver.find_element_by_xpath("//div[@class='a-row icon-farm-wrapper']").find_elements_by_xpath(".//div")
        found=False
        for i in return_ex:
            if(i.get_attribute('data-name')=='RETURNS_POLICY'):
                found=True
                re.append(i.find_element_by_xpath(".//span[1]/div[2]/a").text)
            if(found==False):
                re.append('-')         
    except:
        re.append('-')
        driver.implicitly_wait(1)
    try:
        for a in driver.find_elements_by_id('ftCountdown'):
            exp_delivary.append(a.text)
    except:
        exp_delivary.append('-')
        driver.implicitly_wait(1)
    try:
        for a in driver.find_elements_by_xpath('//span[@class="a-size-medium a-color-success"]'):
            avl.append(a.text)
    except:
        avl.append('-')
        driver.implicitly_wait(1)
        continue
```

```python
# type in search bar
search= driver.find_element_by_xpath("//input[@id='search-input-field']")
search.send_keys("London")
driver.implicitly_wait(5)
```

```python
# Select form dwon menu
select=driver.find_element_by_xpath("/html/body/div[3]/div/div/div[1]/div[1]/div/div[2]/div[4]/div/div[2]/div/div[1]/div/div/ul/li[2]/div").click()
driver.implicitly_wait(2)

# Click on Lets Go
click=driver.find_element_by_xpath("//div[@class='search-button']/button").click()
driver.implicitly_wait(2)
```

```python
# Empty List
hostel_name=[]
distance=[]
reviews=[]
overall_reviews=[]
privates_price=[]
dorms_price=[]
facilities=[]
description=[]
price=[]
ratings=[]

#scrapping data

for a in driver.find_elements_by_xpath("//h2[@class='title title-6']"):
    hostel_name.append(a.text)
for a in driver.find_elements_by_xpath("//span[@class='description']"):
    distance.append(a.text)
for a in driver.find_elements_by_xpath("//div[@class='reviews']"):
    reviews.append(a.text)    
for a in driver.find_elements_by_xpath("//div[@class='keyword']"):
    overall_reviews.append(a.text)    
for a in driver.find_elements_by_xpath("//div[@class='price-col']"):
    price.append(a.text)
    
# Separate privates_price and  dorms_price

for i in range(0,len(price),2):
    privates_price.append(price[i])
for i in range(1,len(price),2):
    dorms_price.append(price[i]) 
    
# Facilities scraping
for a in driver.find_elements_by_xpath("//div[@class='facilities-label facilities']"):
    facilities.append(a.text)
```

```python
# Scapping url for rating and discription
url=[]
for a in driver.find_elements_by_xpath("//h2[@class='title title-6']/a"):
    url.append(a.get_attribute('href'))
    
for i in url:
    try:
        driver.get(i)
        driver.implicitly_wait(2)
    except:
        driver.implicitly_wait(2)
        continue
    for a in driver.find_elements_by_xpath("//div[@class='score orange big']"):
        ratings.append(a.text)  
    else:
        ratings.append('no ratings')
        driver.implicitly_wait(1)
    for a in driver.find_elements_by_xpath("//div[@class='content collapse-content']"):
        description.append(a.text) 
```

```python
Hostel_details=pd.DataFrame({'hostel name':hostel_name[:21],'distance':distance[:21],'reviews':reviews[:21],'overall reviews':overall_reviews[:21],'privates_price':privates_price[:21],
                        'dorms price':dorms_price[:21],'facilities':facilities[:21],'description':description[:21],'price':price[:21], 'ratings':ratings[:21]})

Hostel_details
```
