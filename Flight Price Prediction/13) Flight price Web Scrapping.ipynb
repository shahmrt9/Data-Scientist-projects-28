# Yatra.Com

```python
#connect to web driver
chrome_options=Options()
chrome_options.add_argument("disable--infobars")
chrome_options.add_argument("--incognito")
driver=webdriver.Chrome(r"D:\My Course\chromedriver_win32\chromedriver.exe", options=chrome_options)

#maximize the chrome window
driver.maximize_window() 

# Yatra.com
url="https://www.yatra.com/" 
driver.get(url)
driver.implicitly_wait(4)
```

```python
url=[] # Creating Empty list 
#Getting all the urls from all the location 
for i in range(0,1):
    driver.find_element(By.XPATH,"/html/body/div[2]/div/section/div[2]/div[1]/ul/li[2]").click()
    for i in driver.find_elements(By.XPATH,"//a[@class='VueCarousel-slide items']"):
        url.append(i.get_attribute('href'))
    driver.find_element(By.XPATH,"/html/body/div[2]/div/section/div[2]/div[1]/ul/li[3]").click()
    for i in driver.find_elements(By.XPATH,"//a[@class='VueCarousel-slide items']"):
        url.append(i.get_attribute('href'))
    driver.find_element(By.XPATH,"/html/body/div[2]/div/section/div[2]/div[1]/ul/li[4]").click()
    for i in driver.find_elements(By.XPATH,"//a[@class='VueCarousel-slide items']"):
        url.append(i.get_attribute('href'))
    driver.find_element(By.XPATH,"/html/body/div[2]/div/section/div[2]/div[1]/ul/li[5]").click()
    for i in driver.find_elements(By.XPATH,"//a[@class='VueCarousel-slide items']"):
        url.append(i.get_attribute('href'))
    driver.find_element(By.XPATH,"/html/body/div[2]/div/section/div[2]/div[1]/ul/li[6]").click()
    for i in driver.find_elements(By.XPATH,"//a[@class='VueCarousel-slide items']"):
        url.append(i.get_attribute('href'))    
        
    
print(len(url))
```

```python
# Creating Empty List 
name=[]
depart_time=[]
depart_place=[]
arrive_time=[]
arrive_place=[]
duration=[]
no_stop=[]
price=[]

#Scrapping the details
for a in url:
    driver.get(a)
    try:
        for i in driver.find_elements(By.XPATH,"//div[@class='fs-13 airline-name no-pad col-8']/span"):
            name.append(i.text)
    except:
        for i in driver.find_elements(By.XPATH,"//div[@class='full mb-8 fs-13 airline-name']/span"):
            name.append(i.text)
    try:
        for i in driver.find_elements(By.XPATH,"//div[@class='fs-15 bold time']/div[1]"):
            depart_time.append(i.text)
    except:
        for i in driver.find_elements(By.XPATH,"//div[@class='timing-det bdr-btm v-aligm-t']/div[1]/div[1]/p[1]"):
            depart_time.append(i.text)
    try:
        for i in driver.find_elements(By.XPATH,"//p[1][@class='fs-10 font-lightgrey no-wrap city ellipsis']"):
            depart_place.append(i.text)
    except:                                        
        for i in driver.find_elements(By.XPATH,"//label[1]/div[2]/div[1][@class='grid']/div[1]/p[2]"):
            depart_place.append(i.text)  
    try: 
        for i in driver.find_elements(By.XPATH,"//p[@class='bold fs-15 mb-2 pr time']"):
            arrive_time.append(i.text)
    except:
        for i in driver.find_elements(By.XPATH,"//label[1]/div[2]/div/div[3][@class='i-b col-5 pdd-0 text-left atime']/p[1]"):
            arrive_time.append(i.text)
    try:
        for i in driver.find_elements(By.XPATH,"//p[2][@class='fs-10 font-lightgrey no-wrap city ellipsis']"):
            arrive_place.append(i.text)
    except:
        for i in driver.find_elements(By.XPATH,"//label[1]/div[2]/div/div[3][@class='i-b col-5 pdd-0 text-left atime']/p[2]"):
            arrive_place.append(i.text)
    try:
        for i in driver.find_elements(By.XPATH,"//div[@class='stop-cont pl-13']/p"):
            duration.append(i.text)
    except:
        for i in driver.find_elements(By.XPATH,"//label[1]/div[3]/div[@class='pl-10 stop-cont']/p[1]"):
            duration.append(i.text)
    try:
        for i in driver.find_elements(By.XPATH,"//div[@class='stop-cont pl-13']/div/span"):
            no_stop.append(i.text)
    except:
        for i in driver.find_elements(By.XPATH,"//label[1]/div[3]/div/div[@class=' font-lightgrey fs-11 tipsy i-b']/span"):
            no_stop.append(i.text)
    try:
        for i in driver.find_elements(By.XPATH,"//div[1]/div[4]/div/div[1]/div/label/div/div[2]"):
            price.append(i.text)
    except:
        for i in driver.find_elements(By.XPATH,"//div[2]/div[1]/div[4]/div/div[3]/p"):
            price.append(i.text)
```

```python
# Checking the length 
print(len(name),len(arrive_time),len(duration),len(no_stop),len(arrive_place),len(depart_place),len(depart_time),len(price))
```

```python
# Checking length of all variables
MaxLen=max(len(name),len(arrive_time),len(duration),len(no_stop),len(arrive_place),len(depart_place),len(depart_time),len(price))
MaxLen
```

```python
# Replaceing null values for the price column
price=price+[0]*(MaxLen-len(price))
len(price)
```

```python
# Creating DataFrame
all_flight=pd.DataFrame({'Flight_Name':name,'Departure':depart_time,'Departure_Place':depart_place,'Arrival':arrive_time,'Arrival_Destination':arrive_place,'Total_Stops':no_stop,'Duration':duration,'Price':price})
all_flight
```
