# RATINGS PREDICTION

We have a client who has a website where people write different reviews for technical products. 
Now they are adding a new feature to their website i.e. The reviewer will have to add stars(rating) 
as well with the review. The rating is out 5 stars and it only has 5 options available 1 star, 2 stars, 
3 stars, 4 stars, 5 stars. Now they want to predict ratings for the reviews which were written in the 
past and they don’t have a rating. So, we have to build an application which can predict the rating 
by seeing the review.

```python
import requests
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait, Select
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
import time

from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
```

```python
#connect to web driver
chrome_options=Options()
chrome_options.add_argument("disable--infobars")
chrome_options.add_argument("--incognito")
driver=webdriver.Chrome(r"D:\My Course\chromedriver_win32\chromedriver.exe", options=chrome_options)

#maximize the chrome window
driver.maximize_window() 

# Flipkart webpage by url : https://www.flipkart.com/
url="https://www.flipkart.com/" 
driver.get(url)
driver.implicitly_wait(5)

# closing login pop-up window
Close_button=driver.find_element(By.XPATH, '/html/body/div[2]/div/div/button') 
Close_button.click()

#Entering “key” in search field where “search for products, brands and more” 
key=['laptops', 'Phones', 'Headphones', 'smart watches', 'Professional Cameras', 'Printers', 'Monitors', 'Hometheater', 'Router']

#Create empty list
urls=[]

for i in key:   
    srchBar = driver.find_element(By.XPATH,"//div[@class='_3OO5Xc']/input") #Locating search bar
    srchBar.clear()
    srchBar.send_keys(i)
    
        
    driver.find_element(By.XPATH,"//button[@class='L0Z3Pu']").click() #clicking on search button
    time.sleep(3)
    page = []
    for i in driver.find_elements(By.XPATH,"//nav[@class='yFHi8N']/a"):
        page.append(i.get_attribute('href'))
    for i in page[0:7]:
        driver.get(i)
        time.sleep(3)
        items = driver.find_elements(By.XPATH,"//a[@class='_1fQZEK']")
        for i in items:
            urls.append(i.get_attribute('href'))

print(len(urls))            
```

```python
flip_url1=pd.DataFrame({'URL':urls})
flip_url1
```

```python
def scrap():
    U=driver.find_elements(By.XPATH,"//p[@class='_2-N8zT']/../div")
    for i in U:
        rating.append(i.text)
        driver.implicitly_wait(1)     
    U=driver.find_elements(By.XPATH,"//p[@class='_2-N8zT']")
    for i in U:
        review.append(i.text)
        driver.implicitly_wait(1)            
    U=driver.find_elements(By.XPATH,"//div[@class='t-ZTKy']")
    for i in U:
        story.append(i.text)
        driver.implicitly_wait(1)
```

```python
#Creating empty lits
rating=[]
review=[]
story=[]
#Srapping the data
for a in urls:
    driver.refresh()
    driver.implicitly_wait(3)
    driver.get(a)
    driver.implicitly_wait(3)
    for _ in range(2):
        driver.execute_script("window.scrollBy(0,6000)")
        time.sleep(1)
                   
    # clicking on the all review tab    
    try:
        review_link=driver.find_element(By.XPATH,"//div[@class='_2c2kV-']/following::a")
        for i in review_link:
            link.append(i.get_attribute('href'))
            driver.get(link)
            driver.implicitly_wait(1)
    except:
        pass  
    scrap()
            
    pages = []
    for i in driver.find_elements(By.XPATH,"//nav[@class='yFHi8N']/a"):
        pages.append(i.get_attribute('href'))
    for i in pages[0:17]:
        driver.get(i)
        time.sleep(1)
        scrap()
        
print(len(rating),len(review),len(story))
```
