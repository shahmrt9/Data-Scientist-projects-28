# Cars24 for mumbai location

```python
#connect to web driver
chrome_options=Options()
chrome_options.add_argument("disable--infobars")
chrome_options.add_argument("--incognito")
driver=webdriver.Chrome(r"C:\Users\shaha\Desktop\My Course\DataTrained\chromedriver_win32\chromedriver.exe", options=chrome_options)

# webpage by url :
url="https://www.cars24.com/buy-used-car?sort=P&storeCityId=2378&pinId=400001" 
driver.get(url)

#maximize the chrome window
driver.maximize_window() 
```

```python
#Empty list 
url=[]
# Getting the individual url for used cars
for i in range(0,2):
    try:
        driver.execute_script("window.scrollTo(0,document.body.scrollHeight);") #execute Java script
        driver.implicitly_wait(3)
        for a in driver.find_elements_by_xpath("//div[@class='col-4']/a"):
            url.append(a.get_attribute('href')) 
            continue
           
    except  Exception as e:
        print("Failed",e)

print(url)
len(url)
```

```python
# Creating Empty lists
name=[]
model=[]
fuel=[]
km=[]
Type=[]
price=[]
owner=[]

# Srapping the Data
for i in range(0,2):
    driver.execute_script("window.scrollTo(0,document.body.scrollHeight);") #execute Java script
    driver.implicitly_wait(3)
    for a in driver.find_elements_by_xpath("//h2[@class='_3FpCg']"):
        name.append(a.text)
    for a in driver.find_elements_by_xpath("//p[@class='cvakB']/span"):
        Type.append(a.text)
    for a in driver.find_elements_by_xpath("//ul[@class='bVR0c']/li[3]"):
        fuel.append(a.text)
    for a in driver.find_elements_by_xpath("//ul[@class='bVR0c']/li[1]"):
        km.append(a.text)
    for a in driver.find_elements_by_xpath("//ul[@class='bVR0c']/li[2]"):
        owner.append(a.text.split()[0])
    for a in driver.find_elements_by_xpath("//div[@class='_7udZZ']/span"):
        price.append(a.text)
    for a in driver.find_elements_by_xpath("//p[@class='cvakB']"):
        model.append(a.text)
```

```python
print(len(name),len(Type),len(fuel),len(km),len(owner),len(price),len(model))
```

```python
# Create DataFarme 
cars24=pd.DataFrame({'Name':name,'Model':model,'Fuel':fuel,'Kilometer':km,'Type':Type,'Onwer':owner,'Price':price})
cars24
```

```python
# Save File into CSV formate
cars24.to_csv(r'C:\Users\shaha\Desktop\My Course\1cars24.csv')
```

```python
# Close Window
driver.close()
```

# Cardekho for mumbai location

```python
#connect to web driver
chrome_options=Options()
chrome_options.add_argument("disable--infobars")
chrome_options.add_argument("--incognito")
driver=webdriver.Chrome(r"C:\Users\shaha\Desktop\My Course\DataTrained\chromedriver_win32\chromedriver.exe", options=chrome_options)

# webpage by url :
url="https://www.cardekho.com/used-cars+in+mumbai" 
driver.get(url)

#maximize the chrome window
driver.maximize_window() 
```

```python
#Empty list 
url=[]
# Getting the individual url for used cars
for i in range(0,2):
    driver.execute_script("window.scrollTo(0,document.body.scrollHeight)") #execute Java script
    for a in driver.find_elements_by_xpath("//div[@class='gsc_col-xs-7  carsName']/a"):
        url.append(a.get_attribute('href')) 
        
print(url)
len(url)
```

```python
# Creating Empty lists
name=[]
model=[]
fuel=[]
km=[]
Type=[]
price=[]
owner=[]
# Srapping the Data
for a in driver.find_elements_by_xpath("//div[@class='gsc_col-xs-7  carsName']/a"):
    name.append(a.text)   
for a in driver.find_elements_by_xpath("//div[@class='car_variant']"):
    model.append(a.text)   
for a in driver.find_elements_by_xpath("//div[@class='truncate dotlist-2']/div[2]"):
    fuel.append(a.text)   
for a in driver.find_elements_by_xpath("//div[@class='truncate dotlist-2']/div[1]"):
    km.append(a.text)   
for a in driver.find_elements_by_xpath("//div[@class='truncate dotlist-2']/div[3]"):
    Type.append(a.text)   
for a in driver.find_elements_by_xpath("//div[@class=' price']/span[2]"):
    price.append(a.text)
```

```python
owner=[]
for i in url:
    driver.get(i)
    driver.implicitly_wait(1)
    try:
        for a in driver.find_elements_by_xpath("//div[@id='Overview']/div/div/div/ul/li[6]/div/div[2]"):
            owner.append(a.text)
    except:
        owner.append('NA')
```

```python
# Checking length of all variables
print(len(fuel),len(name),len(model),len(km),len(Type),len(owner),len(price))
```

As we see the lenth of owner is less that the others so, lets fill with 0 and save the data

```python
# Checking length of all variables
maxLen=(max(len(fuel),len(name),len(model),len(km),len(Type),len(owner),len(price)))
```

```python
name=name+[0]*(maxLen-len(name))
price=price+[0]*(maxLen-len(price))
owner=owner+[0]*(maxLen-len(owner))
```

```python
# Again Checking length of all variables
print(len(fuel),len(name),len(model),len(km),len(Type),len(owner),len(price))
```

```python
# Create DataFarme 
cardekho1=pd.DataFrame({'Name':name,'Model':model,'Fuel':fuel,'Kilometer':km,'Type':Type,'Onwer':owner,'Price':price})
cardekho1
```

```python
# Save File into CSV formate
cardekho1.to_csv(r'C:\Users\shaha\Desktop\My Course\cardekho1.csv')
```
