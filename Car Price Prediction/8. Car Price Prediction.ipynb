# Data From OLX

```python
#connect to web driver
chrome_options=Options()
chrome_options.add_argument("disable--infobars")
chrome_options.add_argument("--incognito")
driver=webdriver.Chrome(r"C:\Users\shaha\Desktop\My Course\DataTrained\chromedriver_win32\chromedriver.exe", options=chrome_options)

# webpage by url :
url="https://www.olx.in/items/q-used-cars?isSearchCall=true" 
driver.get(url)

#maximize the chrome window
driver.maximize_window() 
```

```python
#Empty list 
url=[]

# Getting the individual url for used cars
for i in range(0,6):
    try:
        for a in driver.find_elements_by_xpath("//li[@class='EIR5N']/a"):
            url.append(a.get_attribute('href'))
            driver.execute_script("window.scrollTo(0,document.body.scrollHeight);") #execute Java script
            
    except  Exception as e:
        print("Failed", e)
        
    #Click on load more    
    load_more=driver.find_element_by_xpath("//div[@class='JbJAl']/button").click()
    driver.implicitly_wait(2)  
    
print(url)
len(url)
```

```python
# Creating Empty lists
name=[]
model=[]
fuel=[]
km=[]
Type=[]
owner=[]
location=[]
Description=[]
price=[]
image=[]

# Srapping the Data 
for i in url:
    try:
        driver.get(i)
        driver.implicitly_wait(2)
    except:
        driver.implicitly_wait(2)
        continue
    try:
        for a in driver.find_elements_by_xpath("//div[@class='_35xN1']"):
            name.append(a.text)   
    except:
        name.append('NA')
        driver.implicitly_wait(1)
    
    try:
        for a in driver.find_elements_by_xpath("//div[@class='_3tLee']"):
            model.append(a.text)   
    except:
        model.append('NA')
        driver.implicitly_wait(1)
    
    try:
        for a in driver.find_elements_by_xpath("//div[@class='aOxkz']/div[1]"):
            fuel.append(a.text)   
    except:
        fuel.append('NA')
        driver.implicitly_wait(1)
    
    try:
        for a in driver.find_elements_by_xpath("//div[@class='aOxkz']/div[2]"):
            km.append(a.text)   
    except:
        km.append('NA')
        driver.implicitly_wait(1)
    
    try:
        for a in driver.find_elements_by_xpath("//div[@class='aOxkz']/div[3]"):
            Type.append(a.text)   
    except:
        Type.append('NA')
        driver.implicitly_wait(1)
        
    try:
        for a in driver.find_elements_by_xpath("//div[1]/div[@class='_1idEV']/div[2]"):
            owner.append(a.text)   
    except: 
        owner.append('NA')
        driver.implicitly_wait(1)
        
    try:
        for a in driver.find_elements_by_xpath("//div[2]/div[@class='_1idEV']/div[2]"):
            location.append(a.text)   
    except:
        location.append('NA')
        driver.implicitly_wait(1)
    
    try:
        for a in driver.find_elements_by_xpath("//div[@class='_2e_o8']"):
            Description.append(a.text)   
    except:
        Description.append('NA')
        driver.implicitly_wait(1)
    
    try:
        for a in driver.find_elements_by_xpath("//div[@class='_3FkyT']"):
            price.append(a.text)   
    except:
        price.append('NA')
        driver.implicitly_wait(1)
```

```python
# Checking length of all variables
print(len(fuel),len(name),len(model),len(km),len(Type),len(owner),len(location),len(Description),len(price))
```

```python
# Create DataFarme 
olx_car=pd.DataFrame({'Name':name,'Model':model,'Fuel':fuel,'Kilometer':km,'Type':Type,'Onwer':owner,'Price':price,'Location':location,'Description':Description})
olx_car
```

# Data From Cars24 for Pune location

```python
#connect to web driver
chrome_options=Options()
chrome_options.add_argument("disable--infobars")
chrome_options.add_argument("--incognito")
driver=webdriver.Chrome(r"C:\Users\shaha\Desktop\My Course\DataTrained\chromedriver_win32\chromedriver.exe", options=chrome_options)

# webpage by url :
url="https://www.cars24.com/buy-used-car?sort=P&storeCityId=2423&pinId=411001" 
driver.get(url)

#maximize the chrome window
driver.maximize_window() 
```

```python
#Empty list 
url=[]
# Getting the individual url for used cars
for i in range(0,2):
    try:
        driver.execute_script("window.scrollTo(0,document.body.scrollHeight);") #execute Java script
        driver.implicitly_wait(3)
        for a in driver.find_elements_by_xpath("//div[@class='col-4']/a"):
            url.append(a.get_attribute('href')) 
            continue
           
    except  Exception as e:
        print("Failed",e)

print(url)
len(url)
```

```python
# Creating Empty lists
name=[]
model=[]
fuel=[]
km=[]
Type=[]
price=[]
owner=[]

# Srapping the Data
for i in range(0,2):
    driver.execute_script("window.scrollTo(0,document.body.scrollHeight);") #execute Java script
    driver.implicitly_wait(3)
    for a in driver.find_elements_by_xpath("//h2[@class='_3FpCg']"):
        name.append(a.text)
    for a in driver.find_elements_by_xpath("//p[@class='cvakB']/span"):
        Type.append(a.text)
    for a in driver.find_elements_by_xpath("//ul[@class='bVR0c']/li[3]"):
        fuel.append(a.text)
    for a in driver.find_elements_by_xpath("//ul[@class='bVR0c']/li[1]"):
        km.append(a.text)
    for a in driver.find_elements_by_xpath("//ul[@class='bVR0c']/li[2]"):
        owner.append(a.text.split()[0])
    for a in driver.find_elements_by_xpath("//div[@class='_7udZZ']/span"):
        price.append(a.text)
    for a in driver.find_elements_by_xpath("//p[@class='cvakB']"):
        model.append(a.text)
```

```python
print(len(name),len(Type),len(fuel),len(km),len(owner),len(price),len(model))
```

```python
# Create DataFarme 
cars24=pd.DataFrame({'Name':name,'Model':model,'Fuel':fuel,'Kilometer':km,'Type':Type,'Onwer':owner,'Price':price})
cars24
```

```python
# Save File into CSV formate
cars24.to_csv(r'C:\Users\shaha\Desktop\My Course\cars24.csv')
```
